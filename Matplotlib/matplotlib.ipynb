# MATPLOTLIB

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Basic Graph

```python
# basic[x,y] graph
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)

# Add title and labels
plt.title("Our First Graph" ,fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel("X-axis" ,fontdict={'fontname':'Arial',"fontsize":15})
plt.ylabel("Y-axis")
plt.show()
```

# Suppose handling decimal values 

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

#SECOND LINE
#Selecting intervals we want to plot our graph at 
x2=np.arange(0,4.5,0.5)
print(x2)
# printing x^2 insead of y
plt.plot(x2,x2**2, label = "x^2", color="blue" ,linewidth=2, marker="*", markersize=10,
         markeredgecolor="yellow")

# to get int instead of decimals (xticks)
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])
plt.plot(x,y, label="2x", color="red", linewidth=3,linestyle='--',marker="." ,markersize="10",
         markeredgecolor="blue" )
plt.legend()
plt.show()
```

**Plain as well as dashed line**

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

#SECOND LINE
#Selecting intervals we want to plot our graph at 
x2=np.arange(0,4.5,0.5)
print(x2)
# the first four digits as normal line and the remaing as dashed line
plt.plot(x2[:6],x2[:6]**2,label="x^2",color="green")
plt.plot(x2[5:],x2[5:]**2,linestyle="--",color="green")

# to get int instead of decimals (xticks)
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])
plt.plot(x,y, label="2x", color="red", linewidth=3,linestyle='--',marker="." ,markersize="10",
         markeredgecolor="blue" )
plt.legend()
plt.show()
```

# Resize and save your Graph

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

# Resizing the graph
plt.figure(figsize=(4,3), dpi=100) #pixels per inch

#SECOND LINE
x2=np.arange(0,4.5,0.5)
print(x2)
# the first four digits as normal line and the remaing as dashed line
plt.plot(x2[:6],x2[:6]**2,label="x^2",color="green")
plt.plot(x2[5:],x2[5:]**2,linestyle="--",color="green")

# to get int instead of decimals (xticks)
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])
plt.plot(x,y, label="2x", color="red", linewidth=3,linestyle='--',marker="." ,markersize="10",
         markeredgecolor="blue" )
plt.legend()

# Add title and labels
plt.title("Our First Graph" ,fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel("X-axis" ,fontdict={'fontname':'Arial',"fontsize":15})
plt.ylabel("Y-axis")

#Save the graph
plt.savefig("graph1.png",dpi=300)

plt.show()
```

# BAR CHARTS

```python
# A basic bar graph

labels =["Herbivorous","Carnivorous","Omnivorous"]
values=[2,4,9]

# plotting a bar graph
plt.bar(labels,values)

#resizing the graph 
plt.figure(figsize=(1,0.5),dpi=300)

plt.show()
```

# Adding hatch - used to add patterns like stripes,dots,slashesto fill areas such as bar bars in bar charts or areas in pie charts

```python
labels =["Herbivorous","Carnivorous","Omnivorous"]
values=[2,4,9]

# plotting a bar graph
bars =plt.bar(labels,values)

#adding patterns to the bars
bars[0].set_hatch("/")
bars[1].set_hatch("o")
bars[2].set_hatch("*")
#resizing the graph 
plt.figure(figsize=(1,0.5),dpi=300)

plt.show()
```

```python
# Alternate way to add patterns
labels =["Herbivorous","Carnivorous","Omnivorous"]
values=[2,4,9]

# plotting a bar graph
bars =plt.bar(labels,values)

#adding patterns to the bars
pattern=["/","o","*"]
for bar in bars:
    bar.set_hatch(pattern.pop(0))

#resizing the graph 
plt.figure(figsize=(1,0.5),dpi=300)

plt.show()
```

```python
gas = pd.read_csv("gas_prices.csv")

gas
```

```python
plt.title("Gas prices over years in USD")

plt.plot(gas.Year, gas.USA ,label="USA")
plt.plot(gas.Year, gas.Germany ,label="Germany")


#Alternate Way
plt.plot(gas["Year"], gas["South Korea"] ,label="South Korea")
plt.legend()

plt.figure(figsize=(1,0.5),dpi=10)


plt.show()
```

**Plotting gas rate after every three years**

```python
plt.title("Gas prices over years in USD")

plt.plot(gas.Year, gas.USA ,label="USA" ,color="darkblue", marker=".", markeredgecolor="orange")
plt.plot(gas.Year, gas.Germany ,label="Germany", marker=".", markeredgecolor="orange")
plt.plot(gas.Year,gas.Japan,label="japan",color="green" , marker=".",markeredgecolor="orange")

print(gas.Year[::3])

#Labels
plt.xlabel("Year")
plt.ylabel("US Dollars")

#Plot after every three year
plt.xticks(gas.Year[::3])

#Alternate Way
plt.plot(gas["Year"], gas["South Korea"] ,label="South Korea", marker=".", markeredgecolor="orange")
plt.legend()

plt.figure(figsize=(1,0.5),dpi=10)


plt.show()
```

**Plot all the countries**

```python
gas.head()
```

```python
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country],label=country, marker="*" , linewidth=2 , markeredgecolor="black")

plt.xticks(gas.Year[::3])  

plt.title("Gas prices over the year")

plt.xlabel("Year")
plt.ylabel("US Dollars")

plt.legend()

plt.show()
```

**To print certain countries and passing the fontdict dictionary**

```python
plt.title("Gas prices over the years ",fontdict={"fontweight": "bold","fontsize":"20"})

countries_to_look_at = ["UK","Mexico","Italy","Australia"]

for country in gas :
    if country in countries_to_look_at:
        plt.plot(gas.Year,gas[country] ,label=country ,marker=".")

plt.xticks(gas.Year[::3].to_list()+[2011])        

plt.legend()        
```

```python
fifa["Preferred Foot"]
```

```python
# Creating an array of people who prefer their right or left foot 
left = fifa.loc[fifa["Preferred Foot"]== "Left"].count()[0]
print(left) 

right = fifa.loc[fifa["Preferred Foot"]=="Right"].count()[0]
print(right)

plt.title("Preffered foot of fifa players")

#labels 
labels=["left","right"]

#colors
colors=["blue","pink"]

#plot the pie chart
plt.pie([left,right],labels=labels,colors=colors,autopct="%.2f %%") #autppercentage

plt.show()
```

### Second Pie Chart

```python
fifa.Weight
```

```python
## REmove the lbs from the weight 
fifa["Weight"].info()

fifa.Weight = [int(x.strip("lbs")) if type(x)==str else x for x in fifa.Weight]

fifa.Weight[0]
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125 ) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150 ) & (fifa.Weight < 175) ].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175 ) & (fifa.Weight < 200) ].count()[0]
heavy = fifa.loc[fifa.Weight >=200 ].count()[0] 

medium_heavy

#title
plt.title("Weight Distribution of fifa players (in Lbs)")

# changing the color Scheme
plt.style.use('ggplot')
#weights 
weights = [light ,light_medium,medium ,medium_heavy, heavy]
# labels
labels = ["light" , "light_medium","medium" ,"medium_heavy" , "heavy"]
 
#explode - to make the numbers visible 
explode = [.3,.1,0,0,.3]
plt.pie(weights , labels=labels ,autopct="%.2f%%", explode=explode ,pctdistance=0.8)

plt.show()
```

```python
# Using boxplot to comapre two teams 
plt.style.use("default")

plt.figure(figsize=(5,6))


barcelona= fifa.loc[fifa.Club== "FC Barcelona"]["Overall"]
madrid= fifa.loc[fifa.Club== "Real Madrid"]["Overall"]
revs= fifa.loc[fifa.Club== "New England Revolution"]["Overall"]

labels = ["madrid", "barcelona","New England Revolution"]
boxes=plt.boxplot([madrid,barcelona,revs] ,labels=labels ,patch_artist=True ,medianprops={"linewidth" : 2}) # it does not 
#recognixe the facecolor property if patch_artist is not set to true

#Set the edge color
for box in boxes["boxes"]:
    box.set(color="#7FAEDC", linewidth=2)

    # Change fill color
    box.set(facecolor="#CBCBCB")

plt.title("Professional Football Teams Comparison")
plt.ylabel("Fifa Overall Rating")

plt.show()
```
