# MATPLOLIB

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Basic Graph

```python
# basic[x,y] graph
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)

# Add title and labels
plt.title("Our First Graph" ,fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel("X-axis" ,fontdict={'fontname':'Arial',"fontsize":15})
plt.ylabel("Y-axis")
plt.show()
```

```python
# If you want integer values and not decimal 
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])
plt.plot(x,y)

# Add title and labels
plt.title("Our First Graph" ,fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel("X-axis" ,fontdict={'fontname':'Arial',"fontsize":15})
plt.ylabel("Y-axis")

plt.show()
```

**Legend - Exaplin the labels of the plotted Data**

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])
plt.plot(x,y, label="Line 1")
plt.legend()

# Add title and labels
plt.title("Our First Graph" ,fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel("X-axis" ,fontdict={'fontname':'Arial',"fontsize":15})
plt.ylabel("Y-axis")
plt.show()
```

# Plot Attributes

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])
plt.plot(x,y, label="Line 1", color="red", linewidth=3,linestyle='--',marker="." ,markersize="10",
         markeredgecolor="blue" )
plt.legend()
plt.show()
```

# Suppose handling decimal values 

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

#SECOND LINE
#Selecting intervals we want to plot our graph at 
x2=np.arange(0,4.5,0.5)
print(x2)
# printing x^2 insead of y
plt.plot(x2,x2**2, label = "x^2", color="blue" ,linewidth=2, marker="*", markersize=10,
         markeredgecolor="yellow")

# to get int instead of decimals (xticks)
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])
plt.plot(x,y, label="2x", color="red", linewidth=3,linestyle='--',marker="." ,markersize="10",
         markeredgecolor="blue" )
plt.legend()
plt.show()
```

**Plain as well as dashed line**

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

#SECOND LINE
#Selecting intervals we want to plot our graph at 
x2=np.arange(0,4.5,0.5)
print(x2)
# the first four digits as normal line and the remaing as dashed line
plt.plot(x2[:6],x2[:6]**2,label="x^2",color="green")
plt.plot(x2[5:],x2[5:]**2,linestyle="--",color="green")

# to get int instead of decimals (xticks)
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])
plt.plot(x,y, label="2x", color="red", linewidth=3,linestyle='--',marker="." ,markersize="10",
         markeredgecolor="blue" )
plt.legend()
plt.show()
```

# Resize and save your Graph

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

# Resizing the graph
plt.figure(figsize=(4,3), dpi=100) #pixels per inch

#SECOND LINE
x2=np.arange(0,4.5,0.5)
print(x2)
# the first four digits as normal line and the remaing as dashed line
plt.plot(x2[:6],x2[:6]**2,label="x^2",color="green")
plt.plot(x2[5:],x2[5:]**2,linestyle="--",color="green")

# to get int instead of decimals (xticks)
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])
plt.plot(x,y, label="2x", color="red", linewidth=3,linestyle='--',marker="." ,markersize="10",
         markeredgecolor="blue" )
plt.legend()

# Add title and labels
plt.title("Our First Graph" ,fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel("X-axis" ,fontdict={'fontname':'Arial',"fontsize":15})
plt.ylabel("Y-axis")

#Save the graph
plt.savefig("graph1.png",dpi=300)

plt.show()
```

# BAR CHARTS

```python
# A basic bar graph

labels =["Herbivorous","Carnivorous","Omnivorous"]
values=[2,4,9]

# plotting a bar graph
plt.bar(labels,values)

#resizing the graph 
plt.figure(figsize=(1,0.5),dpi=300)

plt.show()
```

# Adding hatch - used to add patterns like stripes,dots,slashesto fill areas such as bar bars in bar charts or areas in pie charts

```python
labels =["Herbivorous","Carnivorous","Omnivorous"]
values=[2,4,9]

# plotting a bar graph
bars =plt.bar(labels,values)

#adding patterns to the bars
bars[0].set_hatch("/")
bars[1].set_hatch("o")
bars[2].set_hatch("*")
#resizing the graph 
plt.figure(figsize=(1,0.5),dpi=300)

plt.show()
```

```python
# Alternate way to add patterns
labels =["Herbivorous","Carnivorous","Omnivorous"]
values=[2,4,9]

# plotting a bar graph
bars =plt.bar(labels,values)

#adding patterns to the bars
pattern=["/","o","*"]
for bar in bars:
    bar.set_hatch(pattern.pop(0))

#resizing the graph 
plt.figure(figsize=(1,0.5),dpi=300)

plt.show()
```
