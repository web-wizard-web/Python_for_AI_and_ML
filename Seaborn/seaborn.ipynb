# SEABORN

**Seaborn is build on top of matplotlib**

```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# getting sns  inbuilt dataset
print(sns.get_dataset_names())
crash_df = sns.load_dataset("car_crashes")

crash_df.head()
```

# DISTRIBUTION PLOTS

### 1 Distribution plot

```python
sns.distplot(crash_df["not_distracted"] ,kde=False,bins=25) #kde=kernel density estimation
```

### JOINT PLOT

```python
sns.jointplot(x="speeding",y="alcohol", data=crash_df,kind="reg") #regression plot
sns.jointplot(x="speeding",y="alcohol", data=crash_df,kind="kde") #kde plot
sns.jointplot(x="speeding",y="alcohol", data=crash_df,kind="hex") #hex plot
```

### KDE PLOT

```python
sns.kdeplot(crash_df["alcohol"])
```

### PAIR PLOT - Relationship accross the entire dataframe numerical values 

```python
sns.pairplot(crash_df)
```

```python
tips_df=sns.load_dataset("tips")
tips_df.head()

sns.pairplot(tips_df,hue="sex", palette="husl") #hue is used to color the
#data points based on a categorical variable
```

### RUGPLOT - dense sticks where amount is most common

```python
sns.rugplot(tips_df["tip"]) #rug plot is used to show the distribution of a single variable
```

# STYLING

```python
sns.set_style("dark") #set style of the plot(whitegrid, darkgrid, dark, white, ticks)

#graph size
plt.figure(figsize=(4,2))

#Changing label and size
sns.set_context("paper",font_scale=1.5) #context is used to set the scale of the plot (paper, notebook, talk, poster)

sns.despine(left=False ,bottom=True) #despine is used to remove the top and right spines (axes) of the plot
sns.jointplot(x="speeding", y="alcohol", data=crash_df, kind="reg")
```

# MATRIX PLOTS

**HEATMAPS - Heatmaps are used to visualize data in a matrix format, where individual values are represented as colors. They are especially useful for:**

**Showing the magnitude of values across two dimensions (e.g., correlation between variables)**

```python
crash_df.head()
crash_df.drop(columns=["abbrev"], inplace=True)
```

```python
plt.figure(figsize=(10,5))
sns.set_style("white")
sns.set_context("paper", font_scale=1.5)

crash_mx= crash_df.corr() #correlation matrix
#correlation matrix is used to show the correlation between the variables
sns.heatmap(crash_mx ,annot=True, cmap="coolwarm")
#annot is used to show the correlation values on the heatmap
#cmap is used to set the color palette of the heatmap
```

```python
# Another way using pivot table
flights = sns.load_dataset("flights")
flights = flights.pivot_table(index="month", columns ="year", values="passengers") #pivot table is used to
# reshape the data
flights.head()
sns.heatmap(flights, cmap="Blues", linecolor="white", linewidths=1)
```

**Cluster Map - It displays the data as a heatmap.It automatically groups (clusters) similar rows and columns together using hierarchical clustering.**
**Dendrograms (tree diagrams) are shown on the top and left to indicate the clustering structure.**

```python
iris = sns.load_dataset("iris")
species = iris.pop("species") #pop is used to remove the column from the dataframe
sns.clustermap(iris)
```

```python
sns.clustermap(flights, standard_scale=1 ,cmap = "Blues") #standard_scale is used to standardize the data
#standardize the data to have mean=0 and standard deviation=1
```

**PAIR GRID**

```python
iris= sns.load_dataset("iris")
iris_g = sns.PairGrid(iris , hue="species") #PairGrid is used to create a grid of subplots
iris_g.map(plt.scatter) #map is used to map the function to the subplots
```

```python
iris_g = sns.PairGrid(iris , hue="species") #PairGrid is used to create a grid of subplots

iris_g.map_diag(plt.hist) #map_diag is used to map the function to the diagonal subplots
iris_g.map_offdiag(plt.scatter) #map_offdiag is used to map the function to the off-diagonal subplots
iris_g.map_upper(plt.scatter) #map_upper is used to map the function to the upper subplots
iris_g.map_lower(sns.kdeplot) #map_lower is used to map the function to the lower subplots
```

```python
# Customizing the plot
iris_g = sns.PairGrid(iris , hue="species" , x_vars=["sepal_length" , "sepal_width"] , y_vars=["petal_length" , "petal_width"]) 
iris_g.map(plt.scatter)
iris_g.add_legend()
```

**FACET GRID**

```python
tips_fg = sns.FacetGrid(tips_df, col="time", row="smoker")
tips_fg.map(plt.hist, "total_bill" ,bins= 10)
```

```python
tips_fg = sns.FacetGrid(tips_df, col="time", row="smoker")
tips_fg.map(plt.scatter, "total_bill" , "tip" )
```

```python
tipd_fg = sns.FacetGrid(tips_df , col="time" , hue="smoker" ,height=4 , aspect=1.3 , col_order=
                        ["Dinner" , "Lunch"] ,palette="Set1") #aspect is used to set the aspect ratio of the plot meaning width/height
tips_fg.map(plt.scatter, "total_bill" , "tip" ,edgecolor="w") #edgecolor is used to set the color of the edge of the data points
tips_fg.add_legend() #add_legend is used to add the legend to the plot
```

```python
att_df = sns.load_dataset("attention")
att_df.head()
att_fg = sns.FacetGrid(att_df , col= "subject" , col_wrap = 5 , height =1.5 ) # col_wrap is used to set the number of columns in the grid
att_fg.map(plt.plot , "solutions" , "score" , marker="o") #marker is used to set the marker of the data points
```
