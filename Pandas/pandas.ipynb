```python
# Get athlethes having Keith as a name 
bios[bios["name"].str.contains("Keith")].head()
```

```python
# Keep case = False to neglect case sensitivity
bios[bios["name"].str.contains("keith|Patrick" ,case=False)]
```

```python
# Find athlethes born in cities that start with a vowel
bios[bios["born_city"].str.contains(r'^[AEIOUaeiou]', na = False)]

# r is raw string prefix in Python . It tells Python not to interpret backlashes as escape
#  sequences in python.

# ^ represrents the startofthe line or word .

```

```python
# Find athlethes name that contains exactlt two vowels
bios[bios["name"].str.count(r'[AEIOUaeiou]')==2 ]
```

```python
# to find names having repedated consecutive letters
bios[bios["name"].str.contains(r'(.)\1' , na=False)]

# . captures the character 
# \1 checks if the next character is the same as the captured one
```

```python
# Find athlethes with names ending with son or sen

bios[bios["name"].str.contains(r'son|sen$' ,case=False ,na=False)]

#$ reprensents the end of word or line
```

```python
# Find athlethes born in a year starting with '19'

bios[bios["born_date"].str.contains(r'^19' ,na=False)]
```

```python
# Find athlethes with names that does not contai any vowels 
bios[bios["name"].str.contains(r'^[^AEIOUaeiou]+$', na=False)]

# ^ start of the word 
# [^AEIOUaeiou]+ match characters that are not vowels
# & end of the word
```

```python
# Find athlethes whose name contains an hyphen or an apostrophe
bios[bios["name"].str.contains(r'[-!]' ,na=False)]
```

```python
# find athlethes with names that start and end with the same letter
bios[bios["name"].str.contains(r'^([a-zA-Z]).*\1$' ,case=False,na=False)]

# *\ - matches any character in between 
# \1$ - matched any character in the end 
```

```python
# find athlethes with a born_city that has exactly 7 characters 
bios[bios["born_city"].str.contains(r'^[a-z A-Z]{7}$' , na=False)]
```

```python
# Find athlethes with names having 2 or more vowels
bios[bios["name"].str.count(r'[AEIOUaeiou]')>=2]
```

```python
bios[bios["born_country"].isin(["USA", "UK" , "GBR"]) & (bios["name"].str.startswith("Keith"))]
```

```python
bios
```

```python
# categorizing them based on their weights
def categorize_weight(row):
    if row["height_cm"]< 165 and row["weight_kg"]<60:
        return "Light Weight"
    elif row["height_cm"]< 185 and row["weight_kg"]<=80:
        return "Medium Weight"
    else:
        return "Heavy Weight"
    
bios["weight_Category"] = bios.apply(categorize_weight, axis=1)    
bios.head()
```

# Advanced Functionality

**Shift - The Shift function in a pandas Series or DataFrame is used to move the values up or down along the axis**

```python
# Foe example comapring the yesterday revenue of coffee to todays 
coffee.head()
```

```python
coffee["Yesterday_revenue"] = coffee["revenue"].shift(2)
# Shifted by 2 along th axis 
```

```python
# Getting the percentage change
coffee["pct_change"] = coffee["revenue"]/ coffee["Yesterday_revenue"]
```

```python
coffee
```

**Rolling() - performing operations over a sliding window of a fixed size over a series or dataframe . for eg :- looking only at 3 values at a time , moving one step forward and computing something from those three values**

```python
latte = coffee[coffee["Coffee Type"]=="Latte"].copy()
latte["3_day"]=latte["Units Sold"].rolling(3).sum()

latte
```
