```python
# Import nescessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
uni = pd.read_csv("timesData.csv")

uni.head()
```

```python
uni.info()
```

```python
# Get all the different countries in country 
uni["country"].unique()
```

```python
# get the csv for the total gdp for the countries
gdp = pd.read_csv("gdp.csv")

# rename the Country to country
gdp.rename(columns={"Country": "country"},inplace=True)

#rename rhe gdp cplumns 
gdp.rename(columns={"GDP (USD Trillions)": "gdp(USD)"},inplace=True)

gdp.head()
```

```python
# combining both datasets (uni and gdp on country)
uni = pd.merge(uni,gdp,on="country", how="inner")
uni.head()
```

```python
# Clean and convert the GDP column to numeric
uni["gdp(USD)"] = uni["gdp(USD)"].replace({"\$": "", ",": ""}, regex=True).astype(float)
uni["gdp(USD)"] = uni["gdp(USD)"] * 1000000000

# Rank the GDP column and add the ranks to a new column
uni["gdp_rank"] = uni["gdp(USD)"].rank(ascending=False)

# Retrieve the country names and their ranks
ranked_countries = uni[["country", "gdp_rank"]].sort_values(by="gdp_rank")
ranked_countries.head()
```

```python
# Rank universities by their scores (assuming 'score' column exists)
uni["uni_rank"] = uni["score"].rank(ascending=False)

# Retrieve the top universities and their countries
top_universities = uni[["country", "uni_rank", "score"]].sort_values(by="uni_rank").head(10)

# Retrieve the top countries by GDP
top_gdp_countries = uni[["country", "gdp_rank", "gdp(USD)"]].sort_values(by="gdp_rank").head(10)

# Merge the two datasets to compare
comparison = pd.merge(top_universities, top_gdp_countries, on="country", how="inner")

# Display the comparison
comparison
```
